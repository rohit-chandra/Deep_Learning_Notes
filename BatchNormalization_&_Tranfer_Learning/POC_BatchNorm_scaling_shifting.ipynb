```python
# normal distribution random values
x = np.random.randn(32)
```

```python
sns.displot(x, kind="kde", fill=True)
```

```python
# check mean and std
x.mean(), x.std()
```

```python
# let's try to shift by 3 and scale it by 5
x_shift = 5*x + 3
```

```python
# mean and std after shifting
x_shift.mean(), x_shift.std()
```

```python
# distribution is shifted
sns.displot(x_shift, kind="kde", fill=True)
```

```python
# to get the original mean from shifted mean 
# x_shift = 5*x + 3 ==> (x_shift - 3)/ 5 = x

x_original = (1/5)*x_shift - 3/5

# let's say if we don't use scaling and use only shifting instead
x_unscaled =  x_shift + 10
```

```python
x_original.mean()
```

```python
#plot x_original and x_shifted on same graph
# red is original, blue is shited, yellow is unscaled
sns.distplot(x_shift, hist=False, rug=True)
sns.distplot(x_original,  hist=False, rug=True)
sns.distplot(x_unscaled,  hist=False, rug=True)

plt.show()
```
